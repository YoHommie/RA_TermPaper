```python
import os
import matplotlib.pyplot as plt

import pandas as pd

folder_path = os.getcwd()
# List of hashfunction
hashfunctions = ['LFSRHash', 'PolynomialHash', 'SingleStagehashing',"StdhashFamily"]
Graphs = ["AugmentedChain1", "CompleteGraph1", "Erdos-Renyi1"]
filePath1 = folder_path + "\\" + hashfunctions[0] + "\\"+Graphs[0]+".csv"
filePath2 = folder_path + "\\" + hashfunctions[0] + "\\"+Graphs[1]+".csv"
filePath3 = folder_path + "\\" + hashfunctions[0] + "\\"+Graphs[2]+".csv"
filePath4 = folder_path + "\\" + hashfunctions[1] + "\\"+Graphs[0]+".csv"
filePath5 = folder_path + "\\" + hashfunctions[1] + "\\"+Graphs[1]+".csv"
filePath6 = folder_path + "\\" + hashfunctions[1] + "\\"+Graphs[2]+".csv"
filePath7 = folder_path + "\\" + hashfunctions[2] + "\\"+Graphs[0]+".csv"
filePath8 = folder_path + "\\" + hashfunctions[2] + "\\"+Graphs[1]+".csv"
filePath9 = folder_path + "\\" + hashfunctions[2] + "\\"+Graphs[2]+".csv"
filePath10 = folder_path + "\\" + hashfunctions[3] + "\\"+Graphs[0]+".csv"
filePath11 = folder_path + "\\" + hashfunctions[3] + "\\"+Graphs[1]+".csv"
filePath12 = folder_path + "\\" + hashfunctions[3] + "\\"+Graphs[2]+".csv"


d1 = pd.read_csv(filePath1)
d2 = pd.read_csv(filePath2)
d3 = pd.read_csv(filePath3)
d4 = pd.read_csv(filePath4)
d5 = pd.read_csv(filePath5)
d6 = pd.read_csv(filePath6)
d7 = pd.read_csv(filePath7)
d8 = pd.read_csv(filePath8)
d9 = pd.read_csv(filePath9)
d10 = pd.read_csv(filePath10)
d11 = pd.read_csv(filePath11)
d12 = pd.read_csv(filePath12)



```

```python
# Plotting the runtime vs number of vertices for each graph and for a hash function in a single figure
plt.figure(figsize=(10, 6))
plt.plot(d1['vertices'], d1['time taken'], label='AugmentedChain1', marker='o')
plt.plot(d2['vertices'], d2['time taken'], label='CompleteGraph1', marker='o')
plt.plot(d3['vertices'], d3['time taken'], label='Erdos-Renyi1', marker='o')
plt.title('LFSRHash')
plt.xlabel('Number of Vertices')
plt.ylabel('Runtime (seconds)')
plt.yscale('log')
plt.legend()
plt.grid()
plt.savefig(folder_path + "\\figures"+"\\LFSRHash.png")
plt.show()
```
